```python
#Introducimos las bibliotecas
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
import matplotlib.dates as mdates
```

```python
#Usamos el DataFrame
df0=pd.read_csv('vid18_180219.dat', delimiter = ',')
```

```python
#Encabezado
df0.head()
```

```python
#Creamos un DataFrame con las columnas de TIMESTAMP y AirTC_Avg
df0=pd.DataFrame({"TIME" : df0.TIMESTAMP,"TEMP" : df0.AirTC_Avg})

df0.head()
```

```python
#Tipos de Variables
df0.dtypes
```

```python
df0['TIME'] = pd.to_datetime(df0.apply(lambda x: x['TIME'], 1))
```

```python
#Construimos una tabla de unidades de frÃ­o acumuladas UF24 por cada dÃ­a completo de datos a partir del 1 de Noviembre del 2018
df0=df0[(df0["TIME"] > "2018-11-1")]
df0.head()
```

```python
#AÃ±adimos columnas para Hora-Dia-Mes-AÃ±o
df0['HORA'] = df0['TIME'].dt.hour
df0['DIA'] = df0['TIME'].dt.day
df0["MES"] = df0["TIME"].dt.month
df0["AÃ‘O"] = df0["TIME"].dt.year
#Borramos columna TIME
f=len(df0)
df0.index = np.arange(0,f)
df0 = df0.drop(["TIME"], 1)
```

```python
#Encabezado
df0.head()
```

```python
df=pd.DataFrame({})
df["TEMP"]=df0.groupby(["AÃ‘O","MES","DIA","HORA"])["TEMP"].transform("mean")
df.head()
```

```python
#Agregamos dos columnas nuevas con las temperaturas MÃ­nima y MÃ¡ximas diarias del periodo de tiempo analizado
df["TMIN"]=df0.groupby(["AÃ‘O","MES","DIA"])["TEMP"].transform("min")
df["TMAX"]=df0.groupby(["AÃ‘O","MES","DIA"])["TEMP"].transform("max")
df["HORA"]=df0.HORA
df["DIA"]=df0.DIA
df["MES"]=df0.MES
df["AÃ‘O"]=df0.AÃ‘O
df.head()
```

```python
#Eliminamos filas duplicadas
df=df.drop_duplicates()
f=len(df)
df.index = np.arange(0,f)
df.head()
```

```python
#Calculamos el indice UF24
count=[]
for i in range (0,len(df)):
    z=df.TEMP[i]
    if (z<=1.4) :
        j=0
    if (z>1.4 and z<=2.4) :
        j=0.5   
    if (z>=2.5 and z<=9.1) :
        j=1
    if (z>=9.2 and z<=12.4) :
        j=0.5  
    if (z>=12.5 and z<=15.9) :
        j=0
    if (z>=16 and z<=18) :
        j=-0.5
    if (z>18) :
        j=-1
    count.append(j)
df["UF24"]=count
```

```python
#Encabezado
df.head()
```

```python
#Contruimos una grÃ¡fica con Matplotlib y mostramos la evoluciÃ³n de las temperaturas mÃ­nima y mÃ¡xima en la misma grÃ¡fica
#Para ello debemos eliminar datos duplicados 
#df1=df
df1=df.drop(["TEMP", "HORA", "UF24"],1)
df1=df1.drop_duplicates()
f=len(df1)
df1.index = np.arange(0,f)
df1.head()



df2=pd.DataFrame({})
df2["day"]=df1.DIA
df2["month"]=df1.MES
df2["year"]=df1.AÃ‘O
df2.index = np.arange(0,f)
df2.head()
j=pd.to_datetime(df2)

df1["FECHA"]=j


#Ya que no tenemos dato duplicados podemos graficar
sns.set(rc={'figure.figsize':(12,8.27)})

df4 = df1[['TMAX', "TMIN"]]
fig=plt.figure(); df4.plot(); plt.legend(loc='lower left')
plt.title("EvoluciÃ³n de las temperaturas mÃ­nima y mÃ¡xima")
plt.ylabel("Temperatura ÂºC")
plt.grid(True)
plt.savefig('EvoTemp_TMINyTMAX',plt=2000)
plt.show()
```

```python
#Construimos una grÃ¡fica con Matplotlib, donde se grafique la acumulaciÃ³n de horas frÃ­o desde el primer dÃ­a (suma de UF24). 
#Para ello debemos eliminar datos duplicados 
df3=df.drop(["TEMP", "TMIN","TMAX"],1)
df3=df3.drop_duplicates()
f=len(df3)
df3.index = np.arange(0,f)
df3.head()

df2=pd.DataFrame({})
df2["day"]=df3.DIA
df2["month"]=df3.MES
df2["year"]=df3.AÃ‘O
df2["hour"]=df3.HORA
df2.index = np.arange(0,f)
df2.head()
j=pd.to_datetime(df2)

df3["FECHA"]=j
df3["SUF"]=df3["UF24"].cumsum()

df4 = df3[["SUF"]]
plt.figure(figsize=(20,19))
fig=plt.figure(); df4.plot(); plt.legend(loc='lower left')
plt.title("La acumulaciÃ³n de horas frÃ­o desde el primer dÃ­a (1 de Noviembre del 2018)")
plt.ylabel("Temperatura ÂºC")
plt.grid(True)
plt.savefig('AcumHorasDeFrio',plt=2000)
plt.show()
```
