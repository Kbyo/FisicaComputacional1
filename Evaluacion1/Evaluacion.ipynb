```python
# Importamos las librerías necesarias para nuestro código
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 13})
```

```python
#Leemos el archivo 
df0 = pd.read_csv("vid18_180219.dat")
```

```python
#Creamos una columna para la fecha como variable date
df0['TIMESTAMP'] = pd.to_datetime(df0.apply(lambda x: x['TIMESTAMP'], 1), dayfirst=True)
```

```python
#seleccionamos solo las columnas que nos interesan
df0 = df0.rename(columns={'TIMESTAMP': 'Fecha','AirTC_Avg':'TempA'})
```

```python
df0=df0[(df0['Fecha']>'2018-11-1')]
```

```python
#Creamos una columna para el número de mes y de año, respectivamente
df0['HORA']=df0['Fecha'].dt.hour
```

```python
df=df0[['Fecha', 'TempA', 'HORA']]
```

```python
df.head()
```

```python
df1 = df.groupby([df0['Fecha'].dt.year,df['Fecha'].dt.month,df['Fecha'].dt.day,df['Fecha'].dt.hour]).mean()
```

```python
df1['Tmin'] = df['TempA'].groupby([df['Fecha'].dt.year,df['Fecha'].dt.day,df['Fecha'].dt.hour]).min()
df1['Tmax'] = df['TempA'].groupby([df['Fecha'].dt.year,df['Fecha'].dt.day,df['Fecha'].dt.hour]).max()
```

```python
df1
```

```python
fig = plt.figure(figsize=(8,5))

df1.groupby(df1['Fecha'].dt.month)['Tmin'].mean().plot(kind = 'line', legend = 'Reverse')
df1.groupby(df1['Fecha'].dt.month)['Tmax'].mean().plot(kind = 'line', legend = 'Reverse')
plt.title('Temperaturas Promedio Máxima y Mínima en Función del Tiempo')
plt.ylabel('Temperatura (°C)')
plt.xlabel('Fecha (Mes)')
```
